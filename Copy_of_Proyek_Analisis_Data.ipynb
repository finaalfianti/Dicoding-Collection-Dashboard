## Exploratory Data Analysis (EDA)

### Explore ...

```python
df = pd.read_csv('day.csv', parse_dates=['dteday'])

print("Informasi Dataset:")
print(df.info())

print("\nStatistik Deskriptif:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

df.fillna(df.median(), inplace=True)

cuaca_group = df.groupby('weathersit')['cnt'].agg(['mean', 'median', 'min', 'max', 'std'])
print("\nPengaruh Cuaca terhadap Penyewaan Sepeda:")
print(cuaca_group)

df['day_type'] = df['weekday'].apply(lambda x: 'Weekend' if x in [0, 6] else 'Weekday')
day_group = df.groupby('day_type')['cnt'].agg(['mean', 'median', 'min', 'max', 'std'])
print("\nPerbedaan Penyewaan Sepeda di Hari Kerja dan Akhir Pekan:")
print(day_group)

correlation = df[['temp', 'atemp', 'hum', 'windspeed', 'cnt']].corr()
print("\nKorelasi antara Variabel Cuaca dan Penyewaan Sepeda:")
print(correlation)

weekday_group = df.groupby('weekday')['cnt'].agg(['mean', 'median', 'min', 'max', 'std'])
print("\nDistribusi Penyewaan Sepeda Berdasarkan Hari dalam Seminggu:")
print(weekday_group)
```

**Insight:**

**Day.csv**

***Pengaruh Cuaca terhadap Penyewaan Sepeda***

- Cuaca yang lebih cerah (weathersit = 1) memiliki rata-rata penyewaan 4,876 sepeda/hari, jauh lebih tinggi dibandingkan cuaca hujan ringan = 1,803 sepeda/hari.

- Penyewaan sepeda menurun saat hujan, dengan minimumnya hanya 22 sepeda pada hari tertentu.

Kesimpulan:
- Cuaca yang lebih buruk berdampak negatif terhadap jumlah penyewaan sepeda.

- Penyedia layanan bisa menawarkan diskon pada hari hujan untuk meningkatkan penyewaan.

***Perbedaan Penyewaan Sepeda di Hari Kerja dan Akhir Pekan***
- Hari kerja (weekday) memiliki rata-rata penyewaan yang lebih tinggi (4,550 sepeda/hari) dibandingkan akhir pekan (4,389 sepeda/hari).

- Namun, variasi pada akhir pekan lebih besar (std = 2,042), artinya jumlah penyewaan lebih tidak stabil pada akhir pekan.

Kesimpulan:
- Penyewaan tetap tinggi baik pada hari kerja maupun akhir pekan, tetapi pada hari kerja mungkin lebih banyak pengguna untuk bekerja, sementara akhir pekan lebih banyak pengguna untuk rekreasi.

- Penyewaan sepeda lebih tinggi pada hari kerja dibandingkan akhir pekan.

- Penyedia layanan bisa menawarkan paket langganan untuk pekerja harian dan paket keluarga untuk akhir pekan.

```python
df_hour = pd.read_csv('hour.csv', parse_dates=['dteday'])

# Menampilkan informasi dataset
print("Informasi Dataset:")
print(df_hour.info())

# Statistik deskriptif data numerik
print("\nStatistik Deskriptif:")
print(df_hour.describe())

# Mengecek missing values
print("\nMissing Values:")
print(df_hour.isnull().sum())

# Mengisi missing values (jika ada) dengan median dari setiap kolom numerik
df_hour.fillna(df_hour.select_dtypes(include=['number']).median(), inplace=True)

# 1. Pengaruh Cuaca terhadap Penyewaan Sepeda (Berdasarkan Kondisi Cuaca)
cuaca_group = df_hour.groupby('weathersit')['cnt'].agg(['mean', 'median', 'min', 'max', 'std'])
print("\nPengaruh Cuaca terhadap Penyewaan Sepeda:")
print(cuaca_group)

# 2. Perbedaan Penyewaan Sepeda di Hari Kerja dan Akhir Pekan
df_hour['day_type'] = df_hour['weekday'].apply(lambda x: 'Weekend' if x in [0, 6] else 'Weekday')
day_group = df_hour.groupby('day_type')['cnt'].agg(['mean', 'median', 'min', 'max', 'std'])
print("\nPerbedaan Penyewaan Sepeda di Hari Kerja dan Akhir Pekan:")
print(day_group)

# 3. Korelasi antara Variabel Cuaca dan Penyewaan Sepeda
correlation = df_hour[['temp', 'atemp', 'hum', 'windspeed', 'cnt']].corr()
print("\nKorelasi antara Variabel Cuaca dan Penyewaan Sepeda:")
print(correlation)

# 4. Distribusi Penyewaan Sepeda Berdasarkan Hari dalam Seminggu
weekday_group = df_hour.groupby('weekday')['cnt'].agg(['mean', 'median', 'min', 'max', 'std'])
print("\nDistribusi Penyewaan Sepeda Berdasarkan Hari dalam Seminggu:")
print(weekday_group)

# 5. Pola Penyewaan Sepeda Berdasarkan Jam dalam Sehari
hourly_group = df_hour.groupby('hr')['cnt'].agg(['mean', 'median', 'min', 'max', 'std'])
print("\nPola Penyewaan Sepeda Berdasarkan Jam dalam Sehari:")
print(hourly_group)
```

**Insight** <br>
**Hour.csv** <br>
***Pola Penyewaan Sepeda Berdasarkan Waktu dalam Sehari***

- Puncak penyewaan terjadi pada jam 8 pagi dan 5 sore, mencerminkan jam sibuk perjalanan kerja/sekolah.

- Jumlah penyewaan relatif rendah di malam hari dan dini hari, dengan titik terendah sekitar jam 3-5 pagi.

- Pada siang hari (sekitar jam 12-14), penyewaan meningkat tetapi tidak setinggi jam sibuk pagi dan sore.

Kesimpulan:
- Penyewaan sepeda mengikuti pola aktivitas harian masyarakat, dengan lonjakan saat jam berangkat dan pulang kerja/sekolah.

***Pengaruh Cuaca terhadap Penyewaan Sepeda***
- Cuaca cerah (weathersit = 1) memiliki rata-rata penyewaan tertinggi (204 sepeda/jam).

- Cuaca mendung atau hujan ringan (weathersit = 2-3) menurunkan jumlah penyewaan secara signifikan.

- Cuaca buruk (weathersit = 4, hujan deras/badai) memiliki rata-rata penyewaan paling rendah (hanya 74 sepeda/jam).

Kesimpulan:
- Semakin buruk kondisi cuaca, semakin sedikit orang yang menyewa sepeda.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana cuaca mempengaruhi jumlah penyewaan sepeda?


```python
# Load dataset Day
df_day = pd.read_csv("day.csv")

# Boxplot berdasarkan kondisi cuaca
plt.figure(figsize=(8, 5))
sns.boxplot(x='weathersit', y='cnt', hue='weathersit', data=df_day, palette='coolwarm', legend=False)
plt.xlabel("Kondisi Cuaca")
plt.ylabel("Jumlah Penyewaan Sepeda")
plt.title("Pengaruh Cuaca terhadap Penyewaan Sepeda (Data Harian)")
plt.xticks(ticks=[0, 1, 2, 3], labels=["Cerah", "Berawan", "Hujan Ringan", "Hujan Lebat"])
plt.show()
```

**Insight:**
- Cuaca yang lebih cerah cenderung memiliki jumlah penyewaan lebih tinggi dibandingkan kondisi berawan atau hujan.

```python
# Load dataset Hour
df_hour = pd.read_csv("hour.csv")

plt.figure(figsize=(12, 6))
sns.lineplot(x='hr', y='cnt', hue='weathersit', data=df_hour, palette='coolwarm')
plt.xlabel("Jam")
plt.ylabel("Jumlah Penyewaan Sepeda")
plt.title("Pola Penyewaan Sepeda per Jam Berdasarkan Kondisi Cuaca")
plt.legend(title="Kondisi Cuaca", labels=["Cerah", "Berawan", "Hujan Ringan", "Hujan Lebat"])
plt.show()
```

**Insight:**
- Pada jam sibuk (pagi & sore), meskipun cuaca buruk, tetap ada penyewaan, tetapi jumlahnya lebih rendah.
- Pada malam hari dan dini hari, hampir tidak ada penyewaan saat hujan.

### Pertanyaan 2: Bagaimana perbedaan penyewaan sepeda di hari kerja dan akhir pekan?

```python
# Load dataset
df_day = pd.read_csv("day.csv")
df_hour = pd.read_csv("hour.csv")

# Mapping nama hari berdasarkan weekday (0 = Minggu, 6 = Sabtu)
days = ["Minggu", "Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu"]
df_day["weekday_name"] = df_day["weekday"].apply(lambda x: days[x])
df_hour["weekday_name"] = df_hour["weekday"].apply(lambda x: days[x])

# ---- 1️⃣ Perbandingan penyewaan sepeda antara hari kerja dan akhir pekan ----
plt.figure(figsize=(8,5))
sns.boxplot(x="workingday", y="cnt", data=df_day, palette=["red", "blue"])
plt.xticks([0, 1], ["Akhir Pekan", "Hari Kerja"])
plt.xlabel("Tipe Hari")
plt.ylabel("Jumlah Penyewaan")
plt.title("Perbedaan Penyewaan Sepeda di Hari Kerja vs Akhir Pekan")
plt.show()

# ---- 2️⃣ Pola penyewaan sepeda per jam untuk hari kerja dan akhir pekan ----
plt.figure(figsize=(12,6))
sns.lineplot(data=df_hour, x="hr", y="cnt", hue="workingday", palette=["red", "blue"])
plt.xlabel("Jam")
plt.ylabel("Jumlah Penyewaan")
plt.title("Pola Penyewaan Sepeda per Jam (Hari Kerja vs Akhir Pekan)")
plt.legend(["Akhir Pekan", "Hari Kerja"])
plt.show()
```

**Insight:**
- Hari kerja memiliki jumlah penyewaan yang lebih tinggi karena aktivitas ke kantor/sekolah.
- Hari kerja memiliki dua puncak (jam 8 pagi & jam 5-6 sore), sesuai dengan jam sibuk.
- Akhir pekan memiliki pola lebih rata tanpa puncak signifikan, menunjukkan bahwa orang menggunakan sepeda lebih bebas.
